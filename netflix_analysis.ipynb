```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("netflix_titles.csv")

print("Dataset Loaded Successfully")
print(data.head())
```

```python
print("\nDataset Information")
print(data.info())

print("\nDataset Shape")
print(data.shape)

print("\nColumn Names")
print(data.columns)
```

```python
print("\nMissing Values in Dataset")

missing_values = data.isnull().sum()
print(missing_values)
```

```python
type_count = data['type'].value_counts()

print("\nMovies vs TV Shows")
print(type_count)
```

```python
# Visualization – Movies vs TV Shows
plt.figure()

data['type'].value_counts().plot(kind='bar')

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

```python
# Top 10 Countries Producing Netflix Content
top_countries = data['country'].value_counts().head(10)

print("\nTop 10 Countries")
print(top_countries)
```

```python
# Visualization – Top Countries
plt.figure()

top_countries.plot(kind='bar')

plt.title("Top Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
```

```python
# Content Released Per Year
release_year = data['release_year'].value_counts().sort_index()

print("\nContent Release Trend")
print(release_year.tail(10))
```

```python
plt.figure()

release_year.plot()

plt.title("Netflix Content Release Trend")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
# Top Genres
top_genres = data['listed_in'].value_counts().head(10)

print("\nTop Genres")
print(top_genres)
```

```python
plt.figure()

top_genres.plot(kind='bar')

plt.title("Top Netflix Genres")
plt.xlabel("Genre")
plt.ylabel("Count")

plt.show()
```
